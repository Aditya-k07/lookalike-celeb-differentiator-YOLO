```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# downloading pre trained yolo for pytorch
model = torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
# detection for images through pretrained model
img = 'https://media.istockphoto.com/id/1480574526/photo/happy-multigenerational-people-having-fun-sitting-on-grass-in-a-public-park.jpg?s=612x612&w=0&k=20&c=iIzSiY2FK9mWTCmV8Ip8zpvXma7f1Qbd-UuKXNJodPg='
```

```python
results = model(img)
print(results)
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
import os
import uuid
import pathlib
labels = ['twinkle khanna','raveena tandon']
number_imgs = 13
root_dir = '.'
for path in pathlib.Path(root_dir).iterdir():
    info = path.stat()
    '''ctime = info.st_ctime
    mtime = info.st_mtime
    if mtime <= ctime:
        date_crated = mtime
    else:
        date_created = ctime
        '''
    file_name,file_extension = os.path.splitext(path)
    if(file_name[0]=='t'):
        imgname = os.path.join(root_dir, labels[0]+'.'+str(uuid.uuid1())+ file_extension)
    else:
        imgname = os.path.join(root_dir, labels[1]+'.'+str(uuid.uuid1())+file_extension)
    os.rename(file_name,imgname)
    print(file_name, file_extension)
```
